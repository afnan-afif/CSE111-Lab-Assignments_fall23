```python
#Task-3
#Write your code here
class RickMote:
  def __init__(self,box=False,channel=0,volume=3):
    self.box=box
    self.channel=channel
    self.volume=volume
    self.status='ON'
  def power(self):
    if self.box==True:
      self.box=False
      self.status='OFF'
    else:
      self.box=True
      self.status='ON'
  def showInfo(self):
    print(f'ID Cable Box Status:\nCable Box is: {self.status}')
    if self.box:
      print(f'channel: {self.channel}\nVolume:{self.volume}')
    else:
      pass
  def changeChannel(self,channel1=2):
    if self.box:
      self.channel_list=[0,2,3,6,7,9]
      if channel1 in self.channel_list:
        self.channel=channel1
      else:
        print(f'Channel does not exist')
    else:
      print(f'Power is turned off. Cannot  Cannot change channel.')
  def changeVolumeLevel(self,volume1=1):
    if self.box:
      self.volume+=volume1
    else:
     print(f'Power is turned off. Cannot change volume.')
# Do not change the following lines of code.
oTV = RickMote()
oTV.power()
print("1.############################")
oTV.showInfo()
print("2.############################")
oTV.changeChannel()
oTV.changeVolumeLevel()
oTV.showInfo()
print("3.############################")
oTV.power()
oTV.showInfo()
print("4.############################")
oTV.power()
oTV.changeVolumeLevel(4)
oTV.changeChannel(3)
oTV.showInfo()
print("5.############################")
oTV.changeVolumeLevel(-2)
oTV.showInfo()
print("6.############################")
oTV.power()
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("7.############################")
oTV.power()
oTV.changeChannel(11)
oTV.showInfo()
```

```python
#Task-5
#Write your code here
class PizzaMachine:
  def __init__(self,p_type='Regular',size=6,toppings=''):
    self.p_type=p_type
    self.size=size
    self.toppings=toppings
    self.spice='Regular'
  def customizePizza(self,*n):
      for i in n:
        if type(i)==list:
          for j in range(len(i)):
            if j==len(i)-1:
              self.toppings+=i[j]
            else:
              self.toppings+=i[j]+','
        else:
          self.spice=i
      if self.spice=='Regular' or self.spice=='Hot' or self.spice=='Super Naga':
        if self.toppings!='':
          return(f'Your {self.size}-inch {self.spice} spicy {self.p_type} Pizza is ready with {self.toppings} toppings. Enjoy!')
        else:
          return('No toppings specified! Can\'t bake pizza.')
      else:
          return ('Sorry! Spice level not allowed. Can\'t bake pizza.')

# Do not change the following lines of code.
pizza1 = PizzaMachine()
order1 = pizza1.customizePizza(["Cheese", "Pepperoni"],"Hot")
print("1################################# ")
print(order1)
print("2================================")
pizza2 = PizzaMachine("Vege")
order2 = pizza2.customizePizza("Super Naga")
print("3#################################")
print(order2)
print("4================================")
pizza3 = PizzaMachine("Chicken Blast",12)
order3 = pizza3.customizePizza(["Mushroom"])
print("5#################################")
print(order3)
print("6================================")
pizza4 = PizzaMachine("Beef Bonanza",16)
order4 = pizza4.customizePizza(["Cheese","Beef kalabhuna"],"Mild")
print("7#################################")
print(order4)
print("8================================")


```

```python
#Task-6
# Write your code here
class Vaccine:
  def __init__(self,brand,country,duration):
    self.brand=brand
    self.country=country
    self.duration=duration
class Person:
  def __init__(self,name,age,occupation='General Citizen'):
    self.name=name
    self.age=age
    self.occupation= occupation
    self.flag1=False
    self.flag2=False

  def pushVaccine(self,n,dosage='1st Dose'):
    if self.age<25 and self.occupation!='Student':
      print(f'Sorry {self.name}, Minimum age for taking vaccines is 25 years now.')
    else:
      if dosage=='1st Dose':
        self.brand=n.brand
        self.country=n.country
        self.duration=n.duration
        self.flag1=True
        print(f'1st dose done for {self.name}')
      elif dosage=='2nd Dose':
        if self.brand==n.brand:
          self.brand=n.brand
          self.country=n.country
          self.duration=n.duration
          self.flag2=True
          print(f'2nd dose done for {self.name}')
        else:
          print(f'Sorry {self.name}, you can’t take 2 different vaccines')
  def showDetail(self):
    print(f'Name: {self.name} Age: {self.age} Type: {self.occupation}\nVaccine name: {self.brand}')
    print(f'1st Dose: Given')
    if self.flag2==True:
      print(f'2nd Dose: Given')
    else:
      print(f'2nd Dose: Please come after {self.duration} days')


# Do not change the following lines of code.
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
