```python
#Task-4
class abcTech:
  def __init__(self,name,designation,department):
    self.name=name
    print(f'Welcome to abcTech, {name}!')
    self.designation=designation
    self.department=department
    self.programming_skills=''
    self.frameworks=''
  def addProgrammingSkills(self,n):
    for i in range(len(n)):
      if i==len(n)-1:
        self.programming_skills+=n[i]
      else:
        self.programming_skills+=', '+n[i]
  def addFrameworks(self,n1):
    for j in range(len(n1)):
      if j==len(n1)-1:
        self.frameworks+=n1[j]
      else:
        self.frameworks+=', '+n1[j]
  def printInfo(self):
    print(f'Name: {self.name}\nDesignation: {self.designation}\nDepartment: {self.department}\nProgramming Skills: {self.programming_skills}\nFrameworks: {self.frameworks}')
  def calculateSalary(self,a,b):
    if b<=144:
      return a

    else:
      c=b-144
      return (a+(c*800))

print('-------------------------')
b1 =abcTech('Tamim Hasan', 'Software Engineer', 'Android Development')
print('-------------------------')
b1.addProgrammingSkills(['Java','Python'])
b1.addProgrammingSkills(['Dart', 'C++'])
b1.addFrameworks(['Express.js','React'])
b1.printInfo()
print('-------------------------')
print(f'Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}')
print('-------------------------')
print('-------------------------')
b2 =abcTech('Jahin Khandoker', 'Senior Developer', 'App Development')
print('-------------------------')
b2.addProgrammingSkills(['Java', 'Dart', 'Swift'])
b2.addFrameworks(['Flutter', 'React Native'])
b2.addFrameworks(['Xamarin'])
b2.printInfo()
print('-------------------------')
print(f'Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}')
print('-------------------------')
```

```python
#Task-6
class Store:
    def __init__(self,balance):
        self.balance=balance
        self.total_items=0
        self.all_items=[]
        self.all_item_dict={}
        print("New branch created!")
    def balance(self):
        return self.balance
    def total_items(self):
        return total_items
    def viewAllItems(self):
        if self.all_items==[]:
            return 'There are no items in your inventory'
        else:
            return self.all_items
    def viewAllItemDetails(self):
        return self.all_item_dict
    def add_item(self,n):
        self.total_items+=1
        self.stock=n[1]
        self.b_value=n[2]
        self.s_value=n[3]
        print(f'Item added: {n[0]}')
        self.balance-=n[2]
        self.all_items.append(n[0])
        self.all_item_dict[n[0]]=''
        d1={'stock': n[1],'buying_price': n[2],'selling_price': n[3]}
        self.all_items_dict.update({n[0]:d1})
    def sell_item(self,x,y):
        if x in self.all_item_dict:
            if y<=self.all_item_dict[x]['stock']:
                g=self.all_item_dict[x]['stock']
                self.balance+=(self.all_item_dict[x]['selling_price'])*y
                self.all_item_dict[x]['stock']-=y
            else:
                print(f'Sorry! {x} is not available at your desired quantity. Currently we have: {g}')

        else:
            print('Item not available')
    def restock_item(self,s,t):
        self.all_item_dict[s]['stock']+=t

#Write Your code here
print('==========================')
branch1 = Store(5000)
print(f'Current Balance: {branch1.balance}')
print(f'Total items: {branch1.total_items}')
branch1.viewAllItems()
branch1.viewAllItemDetails()
print('==========================')
print(f'Current Balance: {branch1.balance}')
branch1.add_item(['ChaCha Noodles', 10, 5, 8])
print(f'Current Balance: {branch1.balance}')
branch1.add_item(['Sparrow Shampoo', 5, 10, 20])
print(f'Current Balance: {branch1.balance}')
print('==========================')
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print('==========================')
print(f'Current Balance: {branch1.balance}\n')
# branch1.sell_item('ChaCha Noodles', 15)
print(f'Current Balance: {branch1.balance}\n')
branch1.viewAllItemDetails()
print()
# branch1.sell_item('ChaCha Noodles', 10)
print()
print(f'Current Balance: {branch1.balance}\n')
branch1.viewAllItemDetails()
print()
print('==========================')
print(f'Current Balance: {branch1.balance}\n')
# branch1.restock_item('ChaCha Noodles', 5)
print()
branch1.viewAllItemDetails()
print()
print(f'Current Balance: {branch1.balance}\n')
print('==========================')
```
