```python
#Task-1
class Spotify:
  def __init__ (self,genre,song1,song2):
    self.genre=genre
    self.song1=song1
    self.song2=song2
  def show_playlist(self):
    return f'Genre: {self.genre}\nSong List: {self.song1},{self.song2}'


#Driver code.
#Write your code here
rock= Spotify("Heavy Metal", "This Fire Burns", "Here I Am")
print(rock.show_playlist())
```

```python
#Task-2
class Exam:
  def __init__(self,exam_type,question,marks_per_question):
    self.et=exam_type
    self.q=question
    self.mpq=marks_per_question

  def detail(self):
    tm=(self.q)*(self.mpq)
    return f"Exam Type: {self.et}\nNumber of questions: {self.q}\nMarks per questions: {self.mpq}\nTotal Marks: {tm}"


#Driver code
#Write your code here
e1 = Exam('Midterm', 2, 10)
print(e1.detail())
print("===========================")
e2 = Exam('Final', 3, 10)
print(e2.detail())
```

```python
#Task-3
class CoffeeMachine:
  def __init__(self,brand_name):
    self.bn=brand_name
    self.ingridients=[]
  def insertIngredients(self,*n):
    for i in n:
      self.ingridients.append(i)
  def getDetails(self):
    temp=''
    for i in range(len(self.ingridients)):
      if i==len(self.ingridients)-1:
        temp+=self.ingridients[i]+'.'
      else:
        temp+=self.ingridients[i]+', '
    return f'Brand Name: {self.bn}\nIngredients: {temp}'



#Driver code
#Write your code here
cm1 = CoffeeMachine("Miyako")
cm1.insertIngredients("Coffee beans", "Milk", "Sugar")
print(cm1.getDetails())
```

```python
#Task-4
#Write your code here
class Dog:
  def __init__(self,breed,age,color):
    self.breed=breed
    self.age=age
    self.color=color
    print(f'=======New Dog Brought!=======')
  def change_info(self,breed,age,color):
    self.breed=breed
    self.age=age
    self.color=color

  def print_info(self):
    print(f"Breed: {self.breed} , Age: {self.age} , Color: {self.color}")

#Driver code
d1 = Dog("Bulldog", 1, "Black")
d1.age= 1.5
d1.change_info("German Shepherd", 0.5, "White")
d1.print_info()


d2 = Dog("Golden Retriever", 2, "Yellow")
d2.print_info()
d2.change_info("Golden Retriever", 1, "White")
d2.print_info()


d2.age=d1.age
print(d2.breed, "is", d2.age, "year(s) old.")
```

```python
#Task-5
# Write your code here
class Student:
  def __init__(self,name,department,gpa):
    self.n,self.id=name.split('_')
    self.dept=department
    self.gpa=gpa
  def calculate_CGPA(self):
    a=0
    for i in self.gpa:
      a+=i*3
    cgpa=a/(len(self.gpa*3))
    self.cgpa=cgpa


  def print_details(self):
    print(f'Name:  {self.n} ID:  {self.id} Department:  {self.dept}')
    print(f'CGPA:{self.cgpa}')
    x=self.cgpa
    if x>3.80:
      print('Your academic standing is \'Highest Distinction\'')
    elif x>3.65:
      print('Your academic standing is \'High Distinction\'')
    elif x>3.50:
      print('Your academic standing is \'Distinction\'')
    elif x>2.00:
      print('Your academic standing is \'Satisfactory\'')
    elif x<2.00:
      print('Sorry, you Cannot graduate')

#Driver code
s1 = Student('Dora_15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu_12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================")
s3 = Student('Bob_13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```
