```python
#TAsk-1
def hospital_fee(**fee):
    max=0
    names=''
    for b in fee.values():
        if max<b:
            max=b
        else:
            pass
    for i in fee.items():
        c,d=i
        if max==d:
            names+=(c+',')
    return (max,names)
max_amount, max_payer = hospital_fee(Neymar = 1000, Dembele = 600, Reus = 500, Bale = 1000)
print(f'Highest fee was {max_amount} tk which was paid by {max_payer}.')
```

```python
#task-2
def is_james_bond(a):
    flag1=False
    flag2=False
    flag3=False
    flag=False
    for i in a:
        if flag1 and flag2 and flag3:
            flag = True
        elif flag2 and i==7:
            flag3=True
        elif flag1 and i==0:
            flag2=True
        elif i==0:
            flag1=True
        else:
            flag = False
    return flag
print(is_james_bond( [1, 2, 4, 0, 0, 7, 5] ))
print(is_james_bond( [1, 7, 2, 0, 4, 5, 0] ))
print(is_james_bond( [1, 0, 2, 0, 4, 7, 5] ))
```

```python
#Task-3
def assign_students_to_sections(*args):
    sum=0
    indx=0
    d={}
    for k in args[0]:
        d[k]=[]
    for i in range(1,len(args)):
        for j in args[i]:
            sum+=ord(j)
        indx=sum%(len(args[0]))
        d[args[0][indx]]+=[args[i]]
        sum=0
        indx=0
    return(d)
print(assign_students_to_sections ('ABCDE', 'Alice','Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
#TAsk-4
def username_generator(first_name,last_name, student_id,middle_name=None):
    first=first_name[0:3:1]
    if middle_name==None:
      middle=''
    else:
      middle=middle_name[0:3:1]
    x=last_name
    last=x[len(x)-3::1]
    y=(str(student_id))
    sid=y[len(y)-4::1]
    return first.upper()+middle+last+'_'+sid
first_name, middle_name, last_name, student_id= input ("First Name:"), input("Middle Name:"), input ("Last Name:"), int (input ("Student ID:"))
print(username_generator (first_name, last_name, student_id))
print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
#Task-5
def key_generator(*a):
    l1=[]
    temp=''
    for i in a:
        temp+=chr(ord(i[0])+32)
        for j in range(-2,-len(i),-1):
            temp+=str(ord(i[j]))
        temp+=chr(ord(i[len(i)-1])-32)
        l1.append(temp)
        temp=''
    return l1
key_list = key_generator("Alex","Bob", "Trudy")
print(key_list)
```

```python
#Task-6
import random
def playRockPaperScissor(rounds):
    usc=0
    csc=0
    p=['rock','paper','scissor']
    for i in range(rounds):
        j=input()
        print(j)
        if j!= 'rock' and j!= 'paper' and j!= 'scissor':
            return '*ERROR*:$ Invalid Input Detected $'
        k=random.choice(p)
        print(f'Computer: {k}')
        if j==k:
            usc+=1
            csc+=1
        else:
            if (j=='rock' and k=='paper') or (j=='paper' and k=='scissor') or (j=='scissor' and k=='rock'):
                csc+=1
            elif (k=='rock' and j=='paper') or (k=='paper' and j=='scissor') or (k=='scissor' and j=='rock'):
                usc+=1
    print(f'Your Score: {usc}')
    print(f'Computer\'s Score: {csc}')
    if usc>csc:
        print('You have won the game!')
    elif usc==csc:
        print('Draw')
    else:
        print('Computer has won the game!')
playRockPaperScissor(int(input("Rounds: ")))
```
